```python
def unique_order(items):
    seen = set()
    result = []
    for x in items:
        if x not in seen:
            seen.add(x)
            result.append(x)
    return result
```

```python
unique_order([3, 1, 2, 3, 2, 4, 1])
```

```python
from collections import Counter

def top_3_words(text):
    top3 = Counter(text.split()).most_common(3)
    return top3
```

```python
top_3_words("cat dog cat bird dog cat")
```

```python
top_3_words("one two three")
```

```python
top_3_words("apple apple banana banana cherry cherry cherry")
```

```python
def merge_sum_dicts(a, b):
    result = {}
    for key in a.keys() | b.keys():
        result[key] = a.get(key, 0) + b.get(key, 0)
    return result
```

```python
merge_sum_dicts({"a": 1, "b": 2}, {"b": 3, "c": 4})
```

```python
merge_sum_dicts({}, {"x": 10})
```

```python
merge_sum_dicts({}, {})
```

```python
def intersection_and_difference(a,b):
    a, b = set(a), set(b)
    return a&b, a-b, b-a
```

```python
intersection_and_difference([1, 2, 3, 4], [3, 4, 5])
```

```python
from collections import defaultdict

def ivert_dict(d):
    result = defaultdict(list)
    for key, value in d.items():
        result[value].append(key)
    return result
```

```python
ivert_dict({1: "a", 2: "b"})
```

```python
ivert_dict({1: "a", 2: "a", 3: "b"})
```

```python
def group_even_odd(l):
    d = {'even': [], 'odd': []}
    for x in l:
        if x%2==0:
            d['even'].append(x)
        else:
            d['odd'].append(x)
    return d
```

```python
group_even_odd([2, 4, 6])
```

```python
def rotate_list(items, k):
    k%=len(items)
    return items[-k:] + items[:-k]
```

```python
rotate_list([1, 2, 3, 4, 5], 2)
```

```python
rotate_list([1, 2, 3], 0)
```

```python
rotate_list([1, 2, 3], 5)
```

```python
def chunks(items, n):
    return [items[i:i + n] for i in range(0, len(items), n)]
```

```python
chunks([1, 2, 3, 4, 5], 2)
```

```python
chunks([1, 2, 3, 4], 1)
```

```python
nums = [2, 7, 11, 15]
target = 9
```

```python
def two_sum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        complement = target-num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
```

```python
two_sum([2, 7, 11, 15], 9)
```

```python
two_sum([3, 2, 4], 6)
```

```python
two_sum([3, 3], 6)
```

```python
two_sum([], 5)
```
