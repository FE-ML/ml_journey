```python
def even_n():
    n = int(input())
    data = []
    for _ in range(n):
        i = int(input())
        if i%2==0:
            data.append(i)
    return data
```

```python
def new_dict(a, b):
    key = a.keys() | b.keys()
    d = {}
    for k in key:
        d[k] = a.get(k,0)+b.get(k,0)
    return d
```

```python
a = {"IT": 100, "HR": 50}
b = {"IT": 40, "Sales": 70}
new_dict(a,b)
```

```python
def union_dict(a,b):
    keys = a.keys() | b.keys()
    d = {}
    for k in keys:
        d[k]=a.get(k,0)+b.get(k,0)
    return d

```

```python
a = {"A": 10, "B": 20}
b = {"B": 5, "C": 15}
union_dict(a,b)
```

```python
from collections import defaultdict
def invert_dict(d):
    result = defaultdict(list)
    for k, v in d.items():
        result[v].append(k)
    return result
```

```python
dic = {"cat": "animal", "dog": "animal", "rose": "plant"}
invert_dict(dic)
```

```python
n = {"a": 1, "b": 2, "c": 1, "d": 2}
invert_dict(n)
```

```python
def sum_k(s, k):
    seen = {}
    result = []
    for i, n in enumerate(s):
        complement = k-n
        if complement in seen:
            result.extend( [seen[complement], i])
        seen[n] = i
    return result
    

nums = [2, 7, 11, 15]
target = 9
sum_k(nums, target)
            
```

```python
def sum_n(s, target):
    seen={}
    result = []
    for i, n in enumerate(s):
        n_2 = target - n
        if n_2 in seen:
            result.extend([i, seen[n_2]])
        seen[n] =i
    return result
```

```python
nums = [3, 1, 4, 8, 6]
target = 10
sum_n(nums, target)
```
