```python
def countdown(n):
    while n>0:
        yield n
        n-=1
```

```python
list(countdown(5))
```

```python
logs = [
    "INFO start",
    "ERROR db failed",
    "INFO stop",
    "ERROR timeout"
]

def error_gen(logs):
    for s in logs:
        if 'ERROR' in s:
            yield s
```

```python
g = error_gen(logs)
print(list(g))
```

```python
def read_csv(filename):
    with open(filename) as f:
        for line in f:
            yield line
```

```python
for row in read_csv('data.csv'):
    print(row)
```

```python
rows = [
    "Anna",
    "",
    "Bob",
    " ",
    "Kate"
]
```

```python
def empty_lines(f):
    for line in f:
        if line.strip():
            yield line
```

```python
for row in empty_lines(rows):
    print(row)
```

```python
def batch_generator(data, batch_size):
    i=0
    while i < len(data):
        yield list(data[i:i+batch_size])
        i+=batch_size
```

```python
list(batch_generator(range(10), 3))
```

```python
samples = [
    ("cat", 0),
    ("dog", 1),
    ("bird", 2),
]
```

```python
def batch_elements(animals):
    i = 0
    while i<len(animals):
        yield list(animals[i:i+2])
        i+=2
```

```python
list(batch_elements(samples))
```
