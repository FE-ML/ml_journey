```python
def safe_divide(a, b):
    try:
        result = a/b
    except ZeroDivisionError:
        print('division by zero')
    else:
        return result
```

```python
safe_divide(10, 2)
# 5.0
```

```python

safe_divide(10, 0)
# "division by zero"
```

```python
def to_int(s):
    try:
        result = int(s)    
    except ValueError:
        print('None')
    else: 
        return result
```

```python
to_int("123")
# 123
```

```python
to_int("abc")
# None
```

```python
values = ["10", "20", "abc", "30"]
```

```python
def correct_sum(s):
    result = 0
    for i in s:
        try:
            result+=int(i)
        except ValueError:
            pass
    return result
```

```python
correct_sum(values)
```

```python
def get_value(data, key):
    try:
        return data[key]
    except KeyError:
        print('missing')
        
```

```python
data = {"a": 1}

get_value(data, "a")
# 1
```

```python

get_value(data, "b")
# "missing"
```

```python
def get_item(items, index):
    try:
        return items[index]
    except IndexError:
        print('None')
```

```python
get_item([1, 2, 3], 1)
# 2
```

```python

get_item([1, 2, 3], 10)
# None
```

```python
def parse_age(text):
    try:
        age = int(text)
    except ValueError:
        print('invalid age')
    else:
        return age
```

```python
parse_age('dsj')
```

```python
parse_age('345')
```

```python
def safe_convert(value):
    try:
        return int(value)
    except ValueError:
        print('None')
    except TypeError:
        print('None')
```

```python
safe_convert('jfj')
```

```python
safe_convert('89')
```

```python
def validate_age(age):
    if age<0:
        raise ValueError('Возраст меньше нуля')
    else: 
        return age
```

```python
validate_age(-4)
```

```python
def check_not_empty(items):
    if items == []:
        raise ValueError('Список пуст')
    else:
        return items
```

```python
check_not_empty([])
```

```python
def validate_probability(p):
    if p>1 or p<0:
        raise ValueError('Вероятность должна быть в диапозоне 0 <= p <= 1')
    else:
        return p
```

```python
validate_probability(9)
```

```python
class DatasetError(Exception):
    pass
```

```python
def load_dataset(path):
    if path.strip():
        return path
    else:
        raise DatasetError('Путь пустой')
```

```python
load_dataset('df')
```

```python
class FileMissingError(DatasetError):
    pass
class  InvalidFormatError(DatasetError):
    pass
```
