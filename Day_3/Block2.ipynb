```python
def make_multiplier(factor):
    
    def inner(num):
        return num*factor
    return inner
```

```python
double = make_multiplier(2)
triple = make_multiplier(3)

print(double(10))
print(triple(10))
```

```python
def make_logger():
    l = []
    def logger(num):
        nonlocal l
        l.append(num)
        return l
    return logger
```

```python
logger = make_logger()
```

```python

print(logger(0.91))
print(logger(0.87))
print(logger(0.95))
```

```python
def make_normalizer(mean, std):

    def normilize(x):
        return (x-mean)/std
    
    return normilize
```

```python
normalize = make_normalizer(
    mean=100,
    std=20
)

print(normalize(120))
```

```python
def make_classifier(threshold):
    def classifier(probability):
        if probability>=threshold:
            return 1
        else:
            return 0
    return classifier
```

```python
clf = make_classifier(0.7)

print(clf(0.8))
print(clf(0.6))
```

```python
def make_cache():
    storage = {}
    def cache(key, value):
        storage[key]=value
        return storage
    return cache
```

```python
cache = make_cache()

cache("a", 1)
cache("b", 2)
```

```python
cache('c', 9)
```

```python
def make_metric(alpha):
    def metric(x):
        return alpha*x
    return metric
```

```python
metric1 = make_metric(0.5)
metric2 = make_metric(2)

print(metric1(10))
print(metric2(10))
```

```python
metric1 = make_metric(0.5)
metric2 = make_metric(2)

print(metric1.__closure__)
print(metric2.__closure__)
```
