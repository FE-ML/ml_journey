```python
def make_early_stopping(patience):
    best_loss = float('inf')
    bad_epochs = 0
    def stop(x):
        nonlocal best_loss, bad_epochs
        if x<best_loss:
            best_loss=x
            bad_epochs=0
        else:
            bad_epochs+=1
        return bad_epochs>=patience
    return stop
```

```python
stop = make_early_stopping(patience=2)
```

```python
stop(1.0)
```

```python
stop(0.9)
```

```python
stop(0.95)
```

```python
stop(0.96)
```

```python
from functools import wraps
```

```python
def log_call(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f'Calling {func}')
        return func(*args, **kwargs)
    return wrapper
```

```python
def log_args(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(args, kwargs)
        return func(*args, **kwargs)
    return wrapper
```

```python
@log_args
def add(a, b): return a + b
```

```python
add(2,5)
```

```python
def show_result(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return f'Result: {result}'
    return wrapper
```

```python
def count_calls(func):
    call = 0
    @wraps(func)
    def wrapper(*args, **kwargs):
        nonlocal call
        call+=1
        print(f'Call #{call}')
        return func(*args, **kwargs)
    return wrapper
```

```python
import time
from functools import wraps
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()-start
        print(end)
        return result
    return wrapper
```

```python
@timer
def slow_fit(): time.sleep(1)
```

```python
slow_fit()
```

```python
def safe(func):
    @wraps(func)
    def wrapper(*args):
        try:
            result = func(*args)
        except ZeroDivisionError:
            print('division by zero')
        else:
            return result
    return wrapper
```

```python
@safe
def divide(a,b):
    return a/b
```

```python
divide(2,1)
```

```python
def only_positive(func):
    @wraps(func)
    def wrapper(*args):
        if args[0]>0:
            return func(*args)
        else:
            raise ValueError
    return wrapper
```

```python
def cache(func):
    seen = {}
    @wraps(func)
    def wrapper(x):
        if x in seen:
            return seen[x]
        result = func(x)
        seen[x] = result
        return result
    return wrapper
        
        
```

```python
@cache
def expensive_transform(x):
    print('Calculating')
    return x*x
```

```python
expensive_transform(4)
```

```python
def repeat(x):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(x):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator
```

```python
@repeat(3)
def augment(): print('aug')
```

```python
augment()
```

```python
def retry(retries, *exceptions):
    def decorator(func):

        @wraps(func)
        def wrapper(*args, **kwargs):
            last_exception=None
            for attempt in range(retries+1):
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    last_exception=e
            return last_exception
        return wrapper
    return decorator
```

```python
def validate_range(low, high):
    def decorator(func):
        @wraps(func)
        def wrapper(x):
            result = func(x)
            if result>high or result < low:
                raise ValueError
            return result
        return wrapper
    return decorator
```

```python
@validate_range(0,1)
def use_probability(p): return p
```

```python
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()-start
        print(end)
        return result
    return wrapper
```

```python
def batch_generator(data, size):
    for i in range(0, len(data), size):
        yield data[i:i+size]
```

```python
@timer
def fit(data, size):
    results = []
    for batch in batch_generator(data, size):
        try:
            results.append(batch)
        except Exception as e:
            print(e)
    return results
```

```python
def retry(retries, exceptions):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            last_exception = None
            for _ in range(retries+1):
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    last_exception=e
            raise last_exception

        return wrapper
    return decorator
```

```python
from pathlib import Path
```

```python
@retry(retries=2, exceptions=(FileNotFoundError,))
def read_file(filename):
    path = Path(filename)
    if not path.exists():
        raise FileExistsError(filename)
    if path.suffix!='.csv' and path.suffix!='.json':
        raise ValueError
    return path
```

```python
read_file('data.csv')
```

```python
read_file('popa.txt')
```
