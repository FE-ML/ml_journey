```python
from functools import wraps
```

```python
def log_call(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        print('Calling function')
        return func(*args, **kwargs)
    return wrapper

@log_call
def greet():
    print("Hello")

greet()
```

```python
def log_args(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f'Arguments: {args}')
        return func(*args, **kwargs)
    return wrapper
```

```python
@log_args
def add(a, b):
    return a + b

add(2, 3)
```

```python
from functools import wraps

def count_calls(func):
    count=0
    @wraps(func)
    def wrapper(*args, **kwargs):
        nonlocal count
        count+=1
        print(f'Call #{count}')
        return func(*args, **kwargs)
    return wrapper
```

```python
@count_calls
def hello():
    print("Hello")

hello()
hello()
hello()
```

```python
import time

def timer(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(round(end-start, 2))
        return result
    return wrapper
```

```python
@timer
def slow():
    time.sleep(2)

slow()
```

```python
def show_result(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        x = func(*args, **kwargs)
        print(f'Result {x}')
        return x
    return wrapper
```

```python
@show_result
def add(a, b):
    return a + b

add(2, 3)
```

```python
def safe(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(e)
        
    return wrapper
```

```python
@safe
def divide(a, b):
    return a / b

divide(10, 0)
```

```python
def deco(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)

    return wrapper
```

```python
@deco
def hello():
    """Greeting"""
    pass

print(hello.__name__)
print(hello.__doc__)
```

```python
def repeat(times):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = None
            for i in range(times):
                
                result = func(*args, **kwargs)
            return result
            
        return wrapper
    return decorator

@repeat(3)
def hello(a,b):
    print("Hello")
    return a+b
print(hello(2,3))
```

```python
def only_positive(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if args[0]>0:
            return func(*args, **kwargs)
        raise ValueError
    return wrapper
```

```python
@only_positive
def square(x):
    return x * x

square(-5)
```

```python
def cache(func):
    cache_data = {}
    @wraps(func)
    def wrapper(x):
        
        if x in cache_data:
            return cache_data[x]
        else:
            result = func(x)
            cache_data[x]=result
            return result
    return wrapper
```

```python
@cache
def square(x):
    print("Calculating")
    return x * x

square(15)
square(15)
square(16)
square(16)
square(26)
square(26)
```

```python
from functools import wraps

def deco(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result

    return wrapper


@deco
def add(a, b):
    """sum"""
    return a + b


print(add.__name__)
print(add.__doc__)
print(add(2, 3))
```
